#### Handling missing values

```python
for col in num_cols:
    df[col].hist(grid=False,  edgecolor="white")
    plt.title(col)
    plt.show()
```

➡️ Most of my numerical columns is skewed so using median to fill missing values

➡️Using Histogram plot to see my numerical columns data and distribution

```python
for col in num_cols:
    df[col]= df[col].fillna(df[col].median())
    
```

```python
for col in num_cols:
    df[col].hist(grid=False,  edgecolor="white")
    plt.title(col)
    plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(15,5))
    sns.countplot(x= df[col], palette='colorblind')
    plt.title(col)
    plt.show()    
```

➡️ Using Proportionate method on Categorical columns to fill missing values

➡️Using Countplot to see my Categorical columns data and distribution

```python
for col in cat_cols:
   proportions= df[col].dropna().value_counts(normalize=True)
   missing_mask = df[col].isnull()
   df.loc[missing_mask, col] = np.random.choice(

    proportions.index, size = missing_mask.sum(), p= proportions.values)
```

### Bivariate analysis

➡️Bivariate analysis is done  to study the relationship between two columns. Here I am  going to compare the relationship between target column and other feature columns. It helps to understand how well my feature and target is corelated or related.

```python
print(num_cols)
len(num_cols)
```

```python
print(cat_cols)
len(cat_cols)
```

```python
target = df['Revenue'], df['ROI']
```

➡️ Analysis between Numerical columns and target->df['Revenue']

```python
features = [col for col in num_cols if col != target]

fig, ax = plt.subplots(len(features), 2, figsize=(12, 4*len(features)))
for i, col in enumerate(features):

    # Scatterplot
    sns.scatterplot(
        x=df[col],
        y=df['Revenue'],
        ax=ax[i,0]
    )

    ax[i,0].set_title(f"{col} vs {'Revenue'}")

    # Boxplot
    sns.boxplot(
        x=df[col],
        ax=ax[i,1]
    )

    ax[i,1].set_title(f"{col} Boxplot")

plt.tight_layout()
plt.show()
```

➡️ Analysis between Categorical columns and target -> df['ROI']

```python
features = [col for col in cat_cols if col != target]
fig,ax = plt.subplots(len(features), 2, figsize = (12,4*len(features)))
for i , col in enumerate(features):

    #boxplot
    sns.boxplot(x=df[col], y=df['ROI'], ax=ax[i,0])
    ax[i,0].set_title(f"{col} vs ROI")

     #Barchart
    sns.barplot(x=df[col], y=df['ROI'], ax=ax[i,1])
    ax[i,1].set_title(f"{col} vs ROI ")
   
plt.tight_layout()
plt.show()

```

## Insights:


### Analyze campaign performance across brands

```python
df['Campaign_Type'].value_counts()
```

➡️ We can see that, all Brands- Tira, Nykaa and Purpelle has generated relatively similar Revenue through all types of Campaign.

➡️ An SEO (Search Engine Optimization) campaign is a strategic, step-by-step plan to 
make your website rank higher on search engines like Google

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df['Campaign_Type'], y=df['Revenue'], hue=df['Brand_Name'], palette='colorblind')
plt.legend(title='Brand Name', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
```

### Identify top-performing and low-performing campaigns

```python
campaign_performance = df.groupby(['Campaign_Type', 'Brand_Name']).agg({'Revenue': 'sum', 'ROI': 'mean', 'Leads':'sum', 'Conversions': 'sum'}).reset_index()
campaign_performance
```

```python
campaign_performance.sort_values(ascending=False, by='Revenue')
```

➡️ SEO, social media, and other campaign types showed comparable performance across all brands, suggesting a balanced marketing strategy.

```python
plt.figure(figsize=(10,5))
sns.barplot(x=campaign_performance['Campaign_Type'], y=campaign_performance['Revenue'], hue=campaign_performance['Brand_Name'], palette='colorblind')
plt.legend(title='Brand Name', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
```

### Relationships between spend, clicks, revenue, and ROI

```python
spend_cost= df.groupby('Brand_Name').agg({'Acquisition_Cost':'sum', 'Clicks':'sum', 'Leads': 'mean', 'Conversions': 'mean'}).reset_index()
spend_cost
```

➡️ As per the graph, all Brands has eaually spent money on acquisition and promotional activities(acquisition cost)

➡️ All Brands has  generated similar number of customers through Clicks(Number of users who clicked on the campaign).

➡️ No of leads(potential customers) generated for all Brands is consistent. (Leads).

➡️ All the 3 Brands has performed similarly into turning all potential customers into buying through their Campaign(Conversions).

```python
plt.figure(figsize=(10,5))
sns.heatmap(spend_cost.set_index('Brand_Name'), fmt= '.1f', annot=True, cmap='coolwarm')
plt.show()
```
